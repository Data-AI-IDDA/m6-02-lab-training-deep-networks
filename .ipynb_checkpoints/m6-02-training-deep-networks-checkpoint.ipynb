### Task 1 — Train the Network

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import numpy as np

device = "cuda" if torch.cuda.is_available() else "cpu"
torch.manual_seed(42)
np.random.seed(42)

device
```

```python
tf = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

train_set = datasets.FashionMNIST(root="./data", train=True, download=True, transform=tf)
val_set = datasets.FashionMNIST(root="./data", train=False, download=True, transform=tf)

train_loader = DataLoader(train_set, batch_size=128, shuffle=True)
val_loader = DataLoader(val_set, batch_size=256, shuffle=False)
```

```python
model = nn.Sequential(
    nn.Flatten(),
    
    nn.Linear(784, 256),
    nn.BatchNorm1d(256),
    nn.ReLU(),
    nn.Dropout(p=0.3),
    
    nn.Linear(256, 128),
    nn.BatchNorm1d(128),
    nn.ReLU(),
    nn.Dropout(p=0.3),
    
    nn.Linear(128, 64),
    nn.BatchNorm1d(64),
    nn.ReLU(),
    
    nn.Linear(64, 10)
).to(device)

print(model)
```

```python
epochs = 15

loss_fn = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-4)
scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs)
```

```python
def evaluate(model, data_loader, loss_fn, device):
    model.eval()
    
    total_loss = 0
    correct = 0
    total = 0
    
    with torch.no_grad():
        for x, y in data_loader:
            x = x.to(device)
            y = y.to(device)
            
            logits = model(x)
            loss = loss_fn(logits, y)
            
            total_loss += loss.item() * x.size(0)
            predictions = logits.argmax(dim=1)
            correct += (predictions == y).sum().item()
            total += y.size(0)
    
    avg_loss = total_loss / total
    accuracy = correct / total
    
    return avg_loss, accuracy
```

```python
train_losses = []
val_losses = []
train_accs = []
val_accs = []

for epoch in range(epochs):
    model.train()
    
    for batch_x, batch_y in train_loader:
        batch_x = batch_x.to(device)
        batch_y = batch_y.to(device)
        
        optimizer.zero_grad()
        logits = model(batch_x)
        loss = loss_fn(logits, batch_y)
        loss.backward()
        optimizer.step()
    
    scheduler.step()
    
    train_loss, train_acc = evaluate(model, train_loader, loss_fn, device)
    val_loss, val_acc = evaluate(model, val_loader, loss_fn, device)
    
    train_losses.append(train_loss)
    val_losses.append(val_loss)
    train_accs.append(train_acc)
    val_accs.append(val_acc)
    
    print(
        f"Epoch {epoch+1}/{epochs} | "
        f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.4f} | "
        f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.4f}"
    )
```

```python
epoch_range = range(1, epochs + 1)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(epoch_range, train_losses, label="Train Loss")
plt.plot(epoch_range, val_losses, label="Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training and Validation Loss")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(epoch_range, train_accs, label="Train Accuracy")
plt.plot(epoch_range, val_accs, label="Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Training and Validation Accuracy")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
best_val_acc = max(val_accs)
best_epoch = val_accs.index(best_val_acc) + 1

print(f"Best validation accuracy: {best_val_acc:.4f}")
print(f"Best epoch: {best_epoch}")
```

### Task 1 — Training Results and Curve Interpretation

The model showed stable learning behaviour throughout the 15 training epochs. The training loss consistently decreased from approximately 0.365 in the first epoch to 0.170 in the final epoch, while the validation loss decreased from about 0.404 to 0.286. This indicates that the neural network successfully learned useful patterns from the Fashion-MNIST dataset over time.

The accuracy curves also showed continuous improvement during training. Training accuracy increased from 86.79% to 93.79%, while validation accuracy improved from 85.43% to 89.84%. The best validation accuracy achieved during training was 89.93% at epoch 14, which is within the expected performance range for this architecture.

From the plotted curves, the training and validation metrics remained relatively close to each other during most of the training process. Although the training accuracy became noticeably higher than the validation accuracy in later epochs, the gap was not extremely large. This suggests that the combination of Batch Normalization, Dropout regularization, weight decay, and the cosine annealing learning rate schedule helped reduce strong overfitting and stabilise the training process.

The validation loss continued to decrease slowly until the final epochs, showing that the model was still improving slightly near the end of training. However, the improvement became smaller after approximately epoch 10, suggesting that the network was beginning to approach convergence.
